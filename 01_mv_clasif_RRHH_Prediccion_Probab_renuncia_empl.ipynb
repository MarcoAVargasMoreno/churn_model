## Conversion de Variables Categoricas a Numericas

### Variables Categoricas
Se crea un dataframe con la variables categoricas que me interesa trabajar, a estas variables hay que volverlas numericas para que el modelo las pueda entender

Las variables categoricas son las que le asigna una etiqueta al cliente en esa variable. Lo clasifica

```
X_cat = employee_df[['BusinessTravel','Department','EducationField','Gender','JobRole','MaritalStatus' ]]
```

```
X_cat
```

Se transforman las variables Categoricas en numericas

Esto se realiza con la libreria: from sklearn.preprocessing import OneHotEncoder

Lo que hace es por ejemplo en la variable gender que tiene los valores Male y Female, las vuelve 2 columnas y deja un 1 en la categoria a la que corresponda

```
from sklearn.preprocessing import OneHotEncoder

# onehotenc --> se crea el objeto hace la transformacion
onehotenc = OneHotEncoder()

X_cat = onehotenc.fit_transform(X_cat).toarray()   # se guarda en un array la transformacion
```

```
X_cat.shape
```

El array que se creo con el OneHotEncoder se lleva a un dataframe

```
X_cat = pd.DataFrame(X_cat)
X_cat.head()
```

### Variables Numericas

Se Crea Dataframe con la variables Numericas del Dataframe Principal

```
# No se trae la columna  o variable objetivo, 'Atrittion'

X_numerical = employee_df[['Age', 'DailyRate', 'DistanceFromHome',	'Education', 'EnvironmentSatisfaction', 'HourlyRate', 'JobInvolvement',	'JobLevel',	'JobSatisfaction',	'MonthlyIncome',	'MonthlyRate',	'NumCompaniesWorked',	'OverTime',	'PercentSalaryHike', 'PerformanceRating',	'RelationshipSatisfaction',	'StockOptionLevel',	'TotalWorkingYears'	,'TrainingTimesLastYear'	, 'WorkLifeBalance',	'YearsAtCompany'	,'YearsInCurrentRole', 'YearsSinceLastPromotion',	'YearsWithCurrManager']]
X_numerical

# otra forma de crear las variables numericas es la siguiente linea
# X_numerical =  employee_df._get_numeric_data()
```

```
X_numerical
```

### Se crea Dataframe final con la union de las variables numericas y las variables categoricas transformadas

Se concatena los dos dataframes con pd.concat()

```
# Axis = 1, se van a unir por filas, ya que el indice es la llave

X_all = pd.concat([X_cat, X_numerical], axis = 1)
X_all
```

### Datos de Entrenamiento y Test

```
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.25)
```

```
X_train.shape
```

```
X_test.shape
```

# TAREA #7: ENTRENAR Y EVALUAR UN CLASIFICADOR UTILIZANDO LA REGRESIÓN LOGÍSTICA

```
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

model = LogisticRegression()
```

Se ajustan los datos de entrenamiento al modelo creado

```
model.fit(X_train, y_train)
```

Se crean las predicciones del modelo entrenado, con la base de testeo

```
y_pred = model.predict(X_test)
```

```
y_pred

# y_pred --> es la salida del modelo

# el 1 en la salida, significa que el empleado abandona la empresa, el cero que no la abandona
```

Medicion de los aciertos (accuracy)

```
from sklearn.metrics import confusion_matrix, classification_report

print("Accuracy {} %".format(100*accuracy_score(y_test, y_pred)))
```

```
# Resultados en el Conjunto de Testing  --> matriz de confusion
cm =confusion_matrix(y_test, y_pred)

# Graficar la matriz
sns.heatmap(cm, annot=True)
```

Se crea la medicion F1 score

```
print(classification_report(y_test, y_pred))
```

# TAREA #8: ENTRENAR Y EVALUR UN CLASIFICADOR UTILIZANDO BOSQUES ALEATORIOS

```
from sklearn.ensemble import RandomForestClassifier

# Se crea el modelo

model = RandomForestClassifier()
```

Se crean las predicciones del modelo entrenado, con la base de testeo

# TAREA #8: ENTRENAR Y EVALUAR UN CLASIFICADOR UTILIZANDO DEEP LEARNING  (Redes Neuronales)

```
import tensorflow as tf
```

Se crea el modelo  --> es un modelo secuencial con la libreria keras

```
model = tf.keras.models.Sequential()

# Se añaden las capas al modelo
# primera capa densa oculta
model.add(tf.keras.layers.Dense(units = 500, activation='relu', input_shape=(50, )))
# segunda capa densa oculta
model.add(tf.keras.layers.Dense(units = 500, activation='relu'))
# tercera capad densa oculta
model.add(tf.keras.layers.Dense(units = 500, activation='relu'))
# Cuarta capa es la de salida --> con la funcion sigmoide , todo lo que salga lo convierte a cero o uno, que es lo que queremos
model.add(tf.keras.layers.Dense(units = 1, activation='sigmoid'))
```

```
model.summary()
```

Se compila  el modelo indicando el optimizador que se va a utilizar , tipicamente alguna variante del gradiente descendente como es el optimizador de Adam  y la funcion de perdidas que se utilizara para medir el error

```
model.compile(optimizer='Adam', loss = 'binary_crossentropy', metrics=['accuracy'])
```

```
# oversampler = SMOTE(random_state=0)
# smote_train, smote_target = oversampler.fit_sample(X_train, y_train)
# epochs_hist = model.fit(smote_train, smote_target, epochs = 100, batch_size = 50)
epochs_hist = model.fit(X_train, y_train, epochs=100, batch_size=50)
```

Se crean las predicciones del modelo entrenado, con la base de testeo

```
# y_pred --> entrega la probabilidad que un empleado se vaya
y_pred = model.predict(X_test)
y_pred
```

Se crea un umbral de precision en la variabl y_test, para compar con los valores reales

```
y_pred = (y_pred > 0.5)
```

Se grafica la perdida y precision del modelo con la historia de la red neuronal en la medida que va realizando las 100 iteraciones

```
epochs_hist.history.keys()
```

```
plt.plot(epochs_hist.history['loss'])
plt.title("Funcion de Perdidas del modelo durante el entrenamiento")
plt.xlabel("Epochs")
plt.ylabel("Error de Entrenamiento")
plt.legend(["Error de Entrenamiento"])
```

```
plt.plot(epochs_hist.history['accuracy'])
plt.title("Tasa de aciertos del modelo durante el entrenamiento")
plt.xlabel("Epochs")
plt.ylabel("Accuracy de Entrenamiento")
plt.legend(["Accuracy de Entrenamiento"])
```

```
# Resultados en el Conjunto de Testing

cm = confusion_matrix(y_test, y_pred)

# se crea mapa de calor con la matriz de confusion
sns.heatmap(cm, annot = True)
```

Se crea la medicion F1_Score

```
print(classification_report(y_test, y_pred))
```
